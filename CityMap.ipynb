```python
import pandas as pd
import folium
from folium.plugins import Fullscreen
```

```python
# Load the CSV file
data_path = 'Desktop/WorkDocs/AGW/data_Marzieh/Charmant/O2_data/Excel/NewDataset.csv'
data = pd.read_csv(data_path)
```

```python
# Filter data to include only rows where 'BarhNo' starts with 'HW' or 'T'
filtered_points = data[data['BarhNo'].str.startswith(('HW', 'T'))]
```

```python
# Create a map centered around the average location of the points
map_karlsruhe = folium.Map(location=[filtered_points['GPSLat'].mean(), filtered_points['GPSLong'].mean()], zoom_start=13)
```

```python
# Add a fullscreen button to the map
Fullscreen(position='topright', title='Expand me', title_cancel='Exit me', force_separate_button=True).add_to(map_karlsruhe)
```

```python
# Add points to the map with labels always shown, with bold text
for idx, point in filtered_points.iterrows():
    marker = folium.Marker(
        location=[point['GPSLat'], point['GPSLong']],
        icon=folium.Icon(icon="info-sign", color="blue")
    )
    # Correctly format popup with HTML for bold text
    popup_content = f"<b>{point['BarhNo']}</b>"
    folium.Popup(popup_content, show=True).add_to(marker)  # Ensure Popup interprets HTML by default
    marker.add_to(map_karlsruhe)
```
